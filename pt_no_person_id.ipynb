```python
import pandas as pd

# Load the three CSV files
file1 = r'C:\Users\SNUH\OneDrive\SNUH BMI Lab\CDM\Holter\pid\holter_pid_20240909.csv'
file2 = r'C:\Users\SNUH\OneDrive\SNUH BMI Lab\CDM\Holter\pid\holter_pid_20240826.csv'
file3 = r'C:\Users\SNUH\OneDrive\SNUH BMI Lab\CDM\Holter\pid\holter_pid_20240828.csv'
file4 = r'C:\Users\SNUH\OneDrive\SNUH BMI Lab\CDM\Holter\pid\pt_no_person_id.csv'

# Read the CSV files
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)

# Select only the 'pt_no' and 'person_id' columns
df1 = df1[['pt_no', 'person_id']]
df2 = df2[['pt_no', 'person_id']]
df3 = df3[['pt_no', 'person_id']]
df4 = df4[['pt_no', 'person_id']]

# Merge the dataframes
merged_df = pd.concat([df1, df2, df3, df4])

# Drop rows with null values in 'pt_no' or 'person_id'
merged_df = merged_df.dropna(subset=['pt_no', 'person_id'])

# Convert 'pt_no' and 'person_id' to integers
merged_df['pt_no'] = merged_df['pt_no'].astype(int)
merged_df['person_id'] = merged_df['person_id'].astype(int)

# Find and count duplicate rows
duplicate_rows = merged_df[merged_df.duplicated(subset=['pt_no', 'person_id'], keep=False)]
duplicate_count = duplicate_rows.shape[0]

# Remove duplicate rows
cleaned_df = merged_df.drop_duplicates(subset=['pt_no', 'person_id'])

# Count the number of rows after removing duplicates
cleaned_count = cleaned_df.shape[0]

print(f"Number of duplicate rows: {duplicate_count}")
print(f"Number of rows after removing duplicates: {cleaned_count}")

# Save the cleaned dataframe to a new CSV file
cleaned_df.to_csv(r'C:\Users\SNUH\OneDrive\SNUH BMI Lab\CDM\Holter\pid\pt_no_person_id.csv', index=False)

print("Cleaned file saved as 'pt_no_person_id.csv'")
```
